```python
import tkinter
from tkinter import ttk
from openpyxl import load_workbook
from tkinter.messagebox import showinfo
import pandas as pd

class Student:
    def __init__(self,std_id,name,std_class,date_of_bir,place_of_bir,contact_name,gender):
        self.std_id = std_id
        self.name = name
        self.std_class = std_class
        self.date_of_bir = date_of_bir
        self.place_of_bir = place_of_bir
        self.contact_name = contact_name
        self.gender = gender

class Application(tkinter.Frame):
    #tạo 1 List rỗng chứa tạm data student

    list_std = []
    
    #tạo 1 biến để lưu data cho widget Treeview
    tree = None
    
    def __init__(self,root):
        super().__init__(root,width=1300,height=660,borderwidth=10,relief='groove')
        self.pack()
        self.pack_propagate(0)
        self.root = root
        self.create_widgets()
        
    def create_widgets(self):
        
        #Nhãn(label) của Ứng Dụng
        app_title_lbl = tkinter.Label(self)
        app_title_lbl['text'] = 'Information Students'
        app_title_lbl['font'] = ('arial',40,'bold')
        app_title_lbl['bg'] = 'lightgrey'
        app_title_lbl['fg'] = 'black'
        app_title_lbl['border'] = 12
        app_title_lbl['relief'] = 'groove'        
        app_title_lbl.pack(fill='x')
        
        #1 Label Frame của phần nhập data detail và các button dùng truy vấn data
        detail_enter_lblframe = tkinter.LabelFrame(self)
        detail_enter_lblframe['text'] = 'Register Infor of Student'
        detail_enter_lblframe['font'] = ('arial',18,'bold')
        detail_enter_lblframe['bg'] = 'lightgrey'
        detail_enter_lblframe['fg'] = 'black'
        detail_enter_lblframe['border'] = 12
        detail_enter_lblframe['relief'] = 'groove'
        detail_enter_lblframe.place(x= 20,y= 100,width= 500,height= 530)
        
        #1.1 Label
        
        #label của std_id
        std_id_lbl = tkinter.Label(detail_enter_lblframe)
        std_id_lbl['text'] = 'Student_ID'
        std_id_lbl['font'] = ('arial',17)
        std_id_lbl['bg'] = 'lightgrey'
        std_id_lbl['fg'] = 'black'
        std_id_lbl.grid(row= 0,column= 0,padx= 20,pady =4)
        
        #Label của std_name
        std_name_lbl = tkinter.Label(detail_enter_lblframe)
        std_name_lbl['text'] = 'Student_Name'
        std_name_lbl['font'] = ('arial',17)
        std_name_lbl['bg'] = 'lightgrey'
        std_name_lbl['fg'] = 'black'
        std_name_lbl.grid(row= 1,column= 0,padx= 20,pady =4)
        
        #Label của std_class
        std_class_lbl = tkinter.Label(detail_enter_lblframe)
        std_class_lbl['text'] = 'Student_Class'
        std_class_lbl['font'] = ('arial',17)
        std_class_lbl['bg'] = 'lightgrey'
        std_class_lbl['fg'] = 'black'
        std_class_lbl.grid(row= 2,column= 0,padx= 20,pady =4)
        
        
        #Label của std_date_of_birth
        date_of_birth_lbl = tkinter.Label(detail_enter_lblframe)
        date_of_birth_lbl['text'] = 'Date of Birth'
        date_of_birth_lbl['font'] = ('arial',17)
        date_of_birth_lbl['bg'] = 'lightgrey'
        date_of_birth_lbl['fg'] = 'black'
        date_of_birth_lbl.grid(row= 3,column= 0,padx= 20,pady =4)
        
        
        #Label của std_place_of_birth
        place_of_birth_lbl = tkinter.Label(detail_enter_lblframe)
        place_of_birth_lbl['text'] = 'Place of Birth'
        place_of_birth_lbl['font'] = ('arial',17)
        place_of_birth_lbl['bg'] = 'lightgrey'
        place_of_birth_lbl['fg'] = 'black'
        place_of_birth_lbl.grid(row= 4,column= 0,padx= 20,pady =4)
        
        #Label của std_Name_of_contact_person
        Name_of_contact_person_lbl = tkinter.Label(detail_enter_lblframe)
        Name_of_contact_person_lbl['text'] = 'Name of contact person'
        Name_of_contact_person_lbl['font'] = ('arial',12)
        Name_of_contact_person_lbl['bg'] = 'lightgrey'
        Name_of_contact_person_lbl['fg'] = 'black'
        Name_of_contact_person_lbl.grid(row= 5,column= 0,padx= 20,pady =4)
        
        
        #2.Label_Frame         
        gender_enter_lblframe = tkinter.LabelFrame(detail_enter_lblframe)
        gender_enter_lblframe['text'] = 'Gender Choice'
        gender_enter_lblframe['font'] = ('arial',14)
        gender_enter_lblframe['bg'] = 'lightgrey'
        gender_enter_lblframe['fg'] = 'black'
        gender_enter_lblframe['border'] = 7
        gender_enter_lblframe['relief'] = 'groove'
        gender_enter_lblframe.place(x= 20,y= 230,width= 450,height= 60)
        
        #2.1 Radio Button
        self.selected_radio = tkinter.StringVar()
        self.gender_radio_btn_male = tkinter.Radiobutton(gender_enter_lblframe)
        self.gender_radio_btn_male['text'] = 'Male'
        self.gender_radio_btn_male['value'] = 'Male'
        self.gender_radio_btn_male['variable'] = self.selected_radio
        self.gender_radio_btn_male['bg'] = 'lightgrey'        
        self.gender_radio_btn_male['font'] = ('arial',12)
        self.gender_radio_btn_male.place(x=180 ,y =0)
        
        
        self.gender_radio_btn_female = tkinter.Radiobutton(gender_enter_lblframe)
        self.gender_radio_btn_female['text'] = 'Female'
        self.gender_radio_btn_female['value'] = 'Female'
        self.gender_radio_btn_female['variable'] = self.selected_radio
        self.gender_radio_btn_female['bg'] = 'lightgrey'        
        self.gender_radio_btn_female['font'] = ('arial',12)
        self.gender_radio_btn_female.place(x=280 ,y =0)
        
        #3. Entry        
        self.std_id_entry = tkinter.Entry(detail_enter_lblframe)
        self.std_id_entry['relief'] = 'groove'
        self.std_id_entry['border'] = 7
        self.std_id_entry['width'] = 40
        self.std_id_entry.grid(row =0, column= 3,padx= 2,pady =2 )   
        
        
        self.std_name_entry = tkinter.Entry(detail_enter_lblframe)
        self.std_name_entry['relief'] = 'groove'
        self.std_name_entry['border'] = 7
        self.std_name_entry['width'] = 40
        self.std_name_entry.grid(row =1, column= 3,padx= 2,pady =2 )      
        
        
        self.std_class_entry = tkinter.Entry(detail_enter_lblframe)
        self.std_class_entry['relief'] = 'groove'
        self.std_class_entry['border'] = 7
        self.std_class_entry['width'] = 40
        self.std_class_entry.grid(row =2, column= 3,padx= 2,pady =2 )        
        
        
        self.date_of_birth_entry = tkinter.Entry(detail_enter_lblframe)
        self.date_of_birth_entry['relief'] = 'groove'
        self.date_of_birth_entry['border'] = 7
        self.date_of_birth_entry['width'] = 40
        self.date_of_birth_entry.grid(row =3, column= 3,padx= 2,pady =2 )         
        
        self.place_of_birth_entry = tkinter.Entry(detail_enter_lblframe)
        self.place_of_birth_entry['relief'] = 'groove'
        self.place_of_birth_entry['border'] = 7
        self.place_of_birth_entry['width'] = 40
        self.place_of_birth_entry.grid(row =4, column= 3,padx= 2,pady =2 )       
        
        
        self.Name_of_contact_person_entry = tkinter.Entry(detail_enter_lblframe)
        self.Name_of_contact_person_entry['relief'] = 'groove'
        self.Name_of_contact_person_entry['border'] = 7
        self.Name_of_contact_person_entry['width'] = 40
        self.Name_of_contact_person_entry.grid(row =5, column= 3,padx= 2,pady =2 )
        
        self.find_id_student_entry = tkinter.Entry(detail_enter_lblframe)
        self.find_id_student_entry['relief'] = 'groove'
        self.find_id_student_entry['border'] = 7
        self.find_id_student_entry['width'] = 38
        self.find_id_student_entry.place(x=20, y= 370 )
        
        self.delete_id_student_entry = tkinter.Entry(detail_enter_lblframe)
        self.delete_id_student_entry['relief'] = 'groove'
        self.delete_id_student_entry['border'] = 7
        self.delete_id_student_entry['width'] = 38
        self.delete_id_student_entry.place(x=20, y= 410 )
        
        
        #Khởi tạo các Button
        
        #1.Button close Ứng Dụng
        quit_btn = tkinter.Button(self)
        quit_btn['text'] = '閉じる'
        quit_btn['command'] = self.root.destroy
        quit_btn.pack(side='bottom')
        
        #2.Button Input to add data đã nhập vào list tạm list_std
        input_reset_display_button = tkinter.Button(detail_enter_lblframe)
        input_reset_display_button['text'] = 'Input,Reset Display'
        input_reset_display_button['relief'] = 'groove'
        input_reset_display_button['bg'] = 'grey'
        input_reset_display_button['width'] = 14
        input_reset_display_button['height'] = 1
        input_reset_display_button['font'] = ('arial',8,'bold')
        input_reset_display_button['command'] = self.add_std
        input_reset_display_button.place(x= 20,y =290)
        
        #3.Button Reset các Entry Detail Information và Save to file
        add_data_to_excel_button = tkinter.Button(detail_enter_lblframe)
        add_data_to_excel_button['text'] = 'Add Data to Excel'
        add_data_to_excel_button['relief'] = 'groove'
        add_data_to_excel_button['bg'] = 'grey'
        add_data_to_excel_button['width'] = 14
        add_data_to_excel_button['height'] = 1
        add_data_to_excel_button['font'] = ('arial',8,'bold')
        add_data_to_excel_button['command'] = self.add_data_to_excel
        add_data_to_excel_button.place(x= 150,y =290)
        
        #4.Button Search all Data from File Excel 
        search_all_button = tkinter.Button(detail_enter_lblframe)
        search_all_button['text'] = 'Search All Data'
        search_all_button['relief'] = 'groove'
        search_all_button['bg'] = 'grey'
        search_all_button['width'] = 14
        search_all_button['height'] = 1
        search_all_button['font'] = ('arial',8,'bold')
        search_all_button['command'] = self.show_data
        search_all_button.place(x= 280,y =290)
        
        #5.Button Choose_data thực hiện việc chọn data của 1 std bất kỳ trong Treeview sau khi nhấn button #4
        choose_data_button = tkinter.Button(detail_enter_lblframe)
        choose_data_button['text'] = 'Choose_data'
        choose_data_button['relief'] = 'groove'
        choose_data_button['bg'] = 'grey'
        choose_data_button['width'] = 14
        choose_data_button['height'] = 1
        choose_data_button['font'] = ('arial',8,'bold')
        choose_data_button['command'] = self.choose_item_update
        choose_data_button.place(x= 280,y =330)
        
        #6.Button reset các widget Entry ở phần trên
        reset_entry_data_button = tkinter.Button(detail_enter_lblframe)
        reset_entry_data_button['text'] = 'Reset_Entry_data'
        reset_entry_data_button['relief'] = 'groove'
        reset_entry_data_button['bg'] = 'grey'
        reset_entry_data_button['width'] = 14
        reset_entry_data_button['height'] = 1
        reset_entry_data_button['font'] = ('arial',8,'bold')
        reset_entry_data_button['command'] = self.reset_entry_data
        reset_entry_data_button.place(x= 150,y =330)
        
        #7.Button update data cua std 
        update_std_id__inf_button = tkinter.Button(detail_enter_lblframe)
        update_std_id__inf_button['text'] = 'Update_std by id'
        update_std_id__inf_button['relief'] = 'groove'
        update_std_id__inf_button['bg'] = 'grey'
        update_std_id__inf_button['width'] = 14
        update_std_id__inf_button['height'] = 1
        update_std_id__inf_button['font'] = ('arial',8,'bold')
        update_std_id__inf_button['command'] = self.update_id_std
        update_std_id__inf_button.place(x= 20,y =330)
        
        #8.Button find data của student theo id
        find_id_student_button = tkinter.Button(detail_enter_lblframe)
        find_id_student_button['text'] = 'find_id_student'
        find_id_student_button['relief'] = 'groove'
        find_id_student_button['bg'] = 'grey'
        find_id_student_button['width'] = 14
        find_id_student_button['height'] = 1
        find_id_student_button['font'] = ('arial',8,'bold')
        find_id_student_button['command'] = self.find_item_by_id_std
        find_id_student_button.place(x= 280,y =370)
        
        delete_id_student_button = tkinter.Button(detail_enter_lblframe)
        delete_id_student_button['text'] = 'delete_id_student'
        delete_id_student_button['relief'] = 'groove'
        delete_id_student_button['bg'] = 'grey'
        delete_id_student_button['width'] = 14
        delete_id_student_button['height'] = 1
        delete_id_student_button['font'] = ('arial',8,'bold')
        delete_id_student_button['command'] = self.delete_item_by_id_std
        delete_id_student_button.place(x= 280,y =410)
    
        
        global tree   # Không có biến global sẽ không tách được việc tạo Tree và việc đem data vào tree
        
        style = ttk.Style() # ttk.Treeview()の外観を変更
        
        style.theme_use('classic') # テーマ(名前)の指定を行う。
        
        tree = ttk.Treeview(self)
        tree['columns'] = (1,2,3,4,5,6,7)
        tree['show'] = 'headings'
        tree.column(1,width=100)
        tree.column(2,width=100)
        tree.column(3,width=100)
        tree.column(4,width=100)
        tree.column(5,width=100)
        tree.column(6,width=100)
        tree.column(7,width=100)
        tree.heading(1,text='std_id')
        tree.heading(2,text='name')
        tree.heading(3,text='std_class')
        tree.heading(4,text='date_of_bir')
        tree.heading(5,text='place_of_bir')
        tree.heading(6,text='contact_name')
        tree.heading(7,text='Gender')
        tree.config(padding=(10,10,10,10))   #set khoảng cách của data đến khung của Treeview 
        tree.config(cursor='mouse')
        scrollbar = ttk.Scrollbar(self, orient=tkinter.VERTICAL, command=tree.yview)
        tree.configure(yscroll=scrollbar.set)
        tree.place(x=530,y=100)
        scrollbar.pack(side=tkinter.RIGHT, fill=tkinter.Y)
   
     #9 Function 
        
        
    #9.1 Function khi Enter button Input,Reset Display sẽ focus cho button Add Data to Excel
 
    def clear_data_treeview(self):
        for item in tree.get_children():
            tree.delete(item)    
        
    def show_data(self):
        self.clear_data_treeview()
        self.file_read()
        std_id=''
        name=''
        std_class=''
        date_of_bir=''
        place_of_bir=''
        contact_name=''
        Gender=''
        i=2       

        while True:            
            std_id = active_sheet.cell(column = 1,row = i).value
            name = active_sheet.cell(column = 2,row = i).value
            std_class = active_sheet.cell(column = 3,row = i).value
            date_of_bir = active_sheet.cell(column = 4,row = i).value
            place_of_bir = active_sheet.cell(column = 5,row =i).value
            contact_name = active_sheet.cell(column = 6,row = i).value
            Gender = active_sheet.cell(column =7,row=i).value
            if std_id == None:
                break
            
            tree.insert("","end",values=(std_id,name,std_class,date_of_bir,place_of_bir,contact_name,Gender))
            i+=1
    
                
    def file_read(self):
        global file
        global book
        global active_sheet    #Biến gloabl này được dùng trong phần nhập data cho Tree
        file = 'Student_information_DB.xlsx'
        
        book=load_workbook(file,data_only=True)
        active_sheet=book.active       
        
     
    def add_std(self):        
        
        #Để thực hiện nhập liên tục nhiều dữ liệu cùng 1 lúc phải xóa các data đã nhập ban đầu
        self.list_std.clear()
        
        #tiến hành lấy data từ các Entry vào các biến        
        std_id= self.std_id_entry.get()
        name= self.std_name_entry.get()
        std_class= self.std_class_entry.get()
        date_of_bir= self.date_of_birth_entry.get()
        place_of_bir= self.place_of_birth_entry.get()
        contact_name= self.Name_of_contact_person_entry.get()
        gender= self.selected_radio.get()
        
        #Thực hiện Save Data nhập từ các Entry vào list_std        
        self.list_std.append([std_id,name,std_class,date_of_bir,place_of_bir,contact_name,gender])
        
        #Thực hiện xóa hiển thị của phần nhập trên Entry
        self.std_id_entry.delete(0,tkinter.END)
        self.std_name_entry.delete(0, tkinter.END)
        self.std_class_entry.delete(0, tkinter.END)
        self.date_of_birth_entry.delete(0, tkinter.END)
        self.place_of_birth_entry.delete(0,tkinter.END)
        self.Name_of_contact_person_entry.delete(0, tkinter.END)
        self.gender_radio_btn_male.deselect()
        self.gender_radio_btn_female.deselect()
        
        showinfo("Student Information Warning", "Please click Button Add Data to Excel for Saving!")        
        
    
    def add_data_to_excel(self):
        df=pd.read_excel(file)
        insert_data=pd.DataFrame(self.list_std,columns=df.columns)
        df=pd.concat([df,insert_data])
        with pd.ExcelWriter(file) as writer:
            df.to_excel(writer,index=False)
        showinfo("Student Information", "Sytem Update Success!")
        
    def choose_item_update(self):
        record_id = tree.focus()
        record_value = tree.item(record_id,'value')
        self.std_id_entry.insert(tkinter.END,record_value[0])
        self.std_name_entry.insert(tkinter.END,record_value[1])
        self.std_class_entry.insert(tkinter.END,record_value[2])
        self.date_of_birth_entry.insert(tkinter.END,record_value[3]) 
        self.place_of_birth_entry.insert(tkinter.END,record_value[4]) 
        self.Name_of_contact_person_entry.insert(tkinter.END,record_value[5])
        if record_value[6] == 'Male':
            self.gender_radio_btn_male.select()          
            
        else:
            self.gender_radio_btn_female.select()
            
        self.find_id_student_entry.insert(tkinter.END,record_value[0])
        self.delete_id_student_entry.insert(tkinter.END,record_value[0])
            
        
            
    def delete_item_by_id_std(self):
        
        delete_id_student = self.delete_id_student_entry.get()       
        
        df =pd.read_excel('Student_information_DB.xlsx')
        list_del=[]
        list_del = df.index[(df['std_id'] == int(delete_id_student))].tolist()
        df = df.drop(index = list_del)
        with pd.ExcelWriter(file) as writer:
            df.to_excel(writer,index=False)            
                
        showinfo("Student Information", "Sytem delete Success!")
        
    def find_item_by_id_std(self):
        student_id = self.find_id_student_entry.get()

        book=load_workbook(file)
        active_sheet=book.active

        sheet_max_row = active_sheet.max_row+2
        sheet_max_column = active_sheet.max_column

        for j in range(1,sheet_max_row+1):
            if active_sheet.cell(row=j,column=1).value == int(student_id):                
                self.std_id_entry.insert(tkinter.END,active_sheet.cell(row = j, column = 1).value)
                self.std_name_entry.insert(tkinter.END,active_sheet.cell(row = j, column =2).value)
                print(f'{self.std_name_entry}')
                self.std_class_entry.insert(tkinter.END,active_sheet.cell(row = j, column =3).value)
                self.date_of_birth_entry.insert(tkinter.END,active_sheet.cell(row = j, column =4).value)
                self.place_of_birth_entry.insert(tkinter.END,active_sheet.cell(row = j, column =5).value)
                self.Name_of_contact_person_entry.insert(tkinter.END,active_sheet.cell(row = j, column =6).value)

                if active_sheet.cell(row = j, column =7).value == 'Male':
                    self.gender_radio_btn_male.select()
                else:
                    self.gender_radio_btn_female.select()

            if j == None:
                break   


        showinfo("Student Information", "Sytem find Success!")
    
    def reset_entry_data(self):
        self.std_id_entry.delete(0,tkinter.END)
        self.std_name_entry.delete(0, tkinter.END)
        self.std_class_entry.delete(0, tkinter.END)
        self.date_of_birth_entry.delete(0, tkinter.END)
        self.place_of_birth_entry.delete(0,tkinter.END)
        self.Name_of_contact_person_entry.delete(0, tkinter.END)
        self.delete_id_student_entry.delete(0, tkinter.END)
        self.find_id_student_entry.delete(0, tkinter.END)        
        self.gender_radio_btn_male.deselect()
        self.gender_radio_btn_female.deselect()
        
    def update_id_std(self):
        df=pd.read_excel(file)        
        update_std_id = self.find_id_student_entry.get()
        list_update = df.index[(df['std_id'] == int(update_std_id))]   # ép kiểu biến update_std_id là int nếu không thì gặp Error
        
        std_id = self.std_id_entry.get()
        std_name = self.std_name_entry.get()
        std_class = self.std_class_entry.get()
        date_of_birth = self.date_of_birth_entry.get()
        place_of_birth = self.place_of_birth_entry.get()
        name_of_contact_person =self.Name_of_contact_person_entry.get()
        gender = self.selected_radio.get()       
        
        df.iloc[list_update,0]= std_id        
        df.iloc[list_update,1]= std_name        
        df.iloc[list_update,2]= std_class        
        df.iloc[list_update,3]= date_of_birth
        df.iloc[list_update,4]= place_of_birth
        df.iloc[list_update,5]= name_of_contact_person
        df.iloc[list_update,6]= gender
        
        with pd.ExcelWriter(file) as writer:
            df.to_excel(writer,index=False)
        
        showinfo("Student Information", "Sytem Update Success!")
        
root = tkinter.Tk()
root.title("Management Information Student")
root.geometry("1350x670")
app = Application(root=root)
app.mainloop()

```
